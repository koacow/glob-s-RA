```python
import pandas as pd
```

```python
disaster_data_path = '/Users/koacow/BOSTON UNIVERSITY Dropbox/Ngoc Duy Khoa Cao/GLOB~S/Data/Natural Disasters/corrected_disaster_data.xlsx'
disaster_df = pd.read_excel(disaster_data_path, engine='openpyxl').dropna(subset=['latitude', 'longitude']).drop(columns=["Latitude", "Longitude"])
disaster_df
```

```python
disaster_df.columns
```

```python
damage_numeric_cols = [
    'Total Deaths',
    'Total Affected',
    'Total Damage, Adjusted (\'000 US$)',
    'Magnitude',
    'Magnitude Scale',
    'No. Injured',
    'No. Affected',
    'No. Homeless',
    'Reconstruction Costs (\'000 US$)',
    'Reconstruction Costs, Adjusted (\'000 US$)',
    'Insured Damage (\'000 US$)', 
    'Insured Damage, Adjusted (\'000 US$)',
    'Total Damage (\'000 US$)'
]
numeric_summary_df = disaster_df[damage_numeric_cols].describe().transpose()
numeric_summary_df
```

```python
numeric_summary_df['Percent Missing'] = 100 * (1 - numeric_summary_df['count'] / len(disaster_df))
numeric_summary_df.to_markdown('./disaster_numeric_damage_summary.md', floatfmt=".2f", index=True)
numeric_summary_df
```
