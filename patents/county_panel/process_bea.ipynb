```python
import pandas as pd
import os
import chardet
```

```python
# 1. Load the BEA county GDP data
CAGDP1 = "/Users/koacow/BOSTON UNIVERSITY Dropbox/Ngoc Duy Khoa Cao/GLOB~S/Data/U.S. County Data/BEA_COUNTY_1969_2023/CAGDP1__ALL_AREAS_2001_2023.csv"
CAGDP1_encoding = 'utf-8'
with open(CAGDP1, 'rb') as file:
    result = chardet.detect(file.read())
    CAGDP1_encoding = result['encoding']
CAGDP1_df = pd.read_csv(CAGDP1, dtype={'GeoFIPS': str}, encoding=CAGDP1_encoding, na_values=['(NA)', '(N/A)'])
CAGDP1_df = CAGDP1_df.rename(columns={
    "GeoFIPS": "FIPS5",
    "GeoName": "geographic_area_name"
})
CAGDP1_df = CAGDP1_df[:-4]
CAGDP1_df
```

```python
# 1.2 Filter the data to only keep rows with 'Real GDP (thousands of chained 2017 dollars)' in the 'Description' column
CAGDP1_df = CAGDP1_df[CAGDP1_df['Description'].str.contains("Real GDP")]
CAGDP1_df
```

```python
# 1.3 Remove the quotes from the FIPS5 column
CAGDP1_df['FIPS5'] = CAGDP1_df['FIPS5'].str.replace('"', '', regex=False)
CAGDP1_df
```

```python
# 1.4 Drop the irrelevant columns
CAGDP1_df = CAGDP1_df.drop(columns=['LineCode','Region', 'Description', 'Unit', 'IndustryClassification', 'TableName'])
years_to_keep = [str(year) for year in range(2001, 2023)]
CAGDP1_df = CAGDP1_df[['FIPS5', 'geographic_area_name'] + years_to_keep]
CAGDP1_df
```

```python
# 2. Conver the table from wide to long format
CAGDP1_df = CAGDP1_df.melt(id_vars=['FIPS5', 'geographic_area_name'], 
                            var_name='year',
                            value_name='rgdp')
CAGDP1_df['year'] = CAGDP1_df['year'].astype(int)
CAGDP1_df['rgdp'] = CAGDP1_df['rgdp'].astype(float)
CAGDP1_df = CAGDP1_df.sort_values(by=['FIPS5', 'year']).reset_index(drop=True)
CAGDP1_df
```

```python
CAGDP1_df['year'].value_counts()
```

```python
CAGDP1_df.isna().sum()
```

```python
# 3.1 Load the BEA county personal income data
CAINC1 = "/Users/koacow/BOSTON UNIVERSITY Dropbox/Ngoc Duy Khoa Cao/GLOB~S/Data/U.S. County Data/BEA_COUNTY_1969_2023/CAINC1__ALL_AREAS_1969_2023.csv"
CAINC1_encoding = 'utf-8'
with open(CAINC1, 'rb') as file:
    result = chardet.detect(file.read())
    CAINC1_encoding = result['encoding']
CAINC1_df = pd.read_csv(CAINC1, dtype={"GeoFIPS": str}, encoding=CAINC1_encoding, na_values=['(NA)', '(N/A)'])
CAINC1_df = CAINC1_df.rename(columns={
    "GeoFIPS": "FIPS5",
    "GeoName": "geographic_area_name"
})
CAINC1_df = CAINC1_df[:-4]
CAINC1_df
```

```python
# 3.2 Filter the data to only keep rows with 'Personal income (thousands of dollars)' in the 'Description' column
CAINC1_df = CAINC1_df[CAINC1_df['Description'].str.contains("Personal income")]
CAINC1_df
```

```python
# 3.3 Remove the quotes from the FIPS5 column
CAINC1_df['FIPS5'] = CAINC1_df['FIPS5'].str.replace('"', '', regex=False)
CAINC1_df
```

```python
# 3.4 Drop the irrelevant columns
CAINC1_df = CAINC1_df.drop(columns=['LineCode','Region', 'Description', 'Unit', 'IndustryClassification', 'TableName'])
years_to_keep = [str(year) for year in range(1980, 2023)]
CAINC1_df = CAINC1_df[['FIPS5', 'geographic_area_name'] + years_to_keep]
CAINC1_df
```

```python
# 4 Conver the table from wide to long format
CAINC1_df = CAINC1_df.melt(id_vars=['FIPS5', 'geographic_area_name'], 
                            var_name='year',
                            value_name='personal_income')
CAINC1_df['year'] = CAINC1_df['year'].astype(int)
CAINC1_df['personal_income'] = CAINC1_df['personal_income'].astype(float)
CAINC1_df = CAINC1_df.sort_values(by=['FIPS5', 'year']).reset_index(drop=True)
CAINC1_df
```

```python
CAINC1_df['year'].value_counts()
```

```python
CAINC1_df.isna().sum()
```

```python
# 5. Merge the two dataframes on FIPS5 and year
merged_bea_df = pd.merge(CAGDP1_df, CAINC1_df, on=['FIPS5', 'year'], how='outer', suffixes=(None, '_y'))
merged_bea_df = merged_bea_df.drop(columns=[col for col in merged_bea_df.columns if col.endswith('_y')])
merged_bea_df
```
