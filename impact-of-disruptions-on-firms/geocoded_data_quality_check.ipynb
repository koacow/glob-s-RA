```python
import pandas as pd
```

```python
df = pd.read_csv('./data/raw/firm_ids_and_cities.csv', encoding='latin1')
df
```

```python
import gc
import os
```

```python
estb_path = '/Users/koacow/BOSTON UNIVERSITY Dropbox/Ngoc Duy Khoa Cao/Climate Risk and Labor Market/RAIS data/firms and cities/estb/treated'

merged_df = None
for f in sorted(os.listdir(estb_path)):
    if f.startswith('estb_2') and f.endswith('.csv'):
        temp_df = pd.read_csv(os.path.join(estb_path, f), encoding='latin1')
        temp_df = pd.merge(temp_df, df, how='inner', on=['cnpj_cei', 'year'])
        if merged_df is None:
            merged_df = temp_df
        else:
            merged_df = pd.concat([merged_df, temp_df], ignore_index=True)
        del temp_df
        gc.collect()

merged_df
```

```python
merged_df['vinculos_ativos'].describe()
```

```python
large_companies_df = None

for f in sorted(os.listdir(estb_path)):
    if f.startswith('estb_2') and f.endswith('.csv'):
        temp_df = pd.read_csv(os.path.join(estb_path, f), encoding='latin1')
        temp_df = pd.merge(temp_df, df, how='inner', on=['cnpj_cei', 'year'])
        temp_df = temp_df[temp_df['vinculos_ativos'] >= 100]
        if large_companies_df is None:
            large_companies_df = temp_df
        else:
            large_companies_df = pd.concat([large_companies_df, temp_df], ignore_index=True)
        del temp_df
        gc.collect()

large_companies_df
```

```python
large_companies_df = large_companies_df.merge(df, how='inner', on=['cnpj_cei', 'year'])
assert large_companies_df.shape[0] == merged_df.shape[0], "The number of records with 100+ employees should be exactly the same as the number of records with geocoded data"
print("The number of records with 100+ employees is the same as the number of records with geocoded data.")
```
